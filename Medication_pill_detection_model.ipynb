```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data= "/content/drive/MyDrive/PillDetection.v1i.yolov11/data.yaml", epochs=100, imgsz=640, project="/content/drive/MyDrive/PillDetection.v1i.yolov11",
    name="pill_run1")
```

```python
from ultralytics import YOLO

# Load from your saved weights
model = YOLO("/content/drive/MyDrive/PillDetection.v1i.yolov11/pill_run1/weights/last.pt")

# Resume training (resume=True automatically uses the last training config)
model.train(resume=True)
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO("/content/drive/MyDrive/PillDetection.v1i.yolov11/pill_run1/weights/best.pt")  # load a fine-tuned model

# Inference using the model
results = model.predict("https://ultralytics.com/assets/medical-pills-sample.jpg")
```
